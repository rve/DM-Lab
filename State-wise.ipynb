```python
#group the mental health services data by year and state(LST)
grouped_ys = dh.groupby(['YEAR','LST'], sort=True).agg({'LST': len}).rename(columns={'LST':'NProviders'}).reset_index()
grouped_ys
```

```python
#top 10 from 2010
g2010 = grouped_ys.groupby(['YEAR']).get_group(2010).reset_index()[['LST','NProviders']].sort_values(by=['NProviders'], ascending=False).head(10)
g2010
```

```python
#function for labeling
def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')
```

```python
width = 0.50 

fig, ax = plt.subplots()
serv10 = ax.bar(np.arange(10), g2010['NProviders'], width, color='r')

ax.set_xticks(np.arange(10) + width / 2)
ax.set_xticklabels(('CA','NY','PA','IL','OH','FL','MI','TX','MA','IN'))

ax.set_ylabel('Number of hospital/service')
ax.set_xlabel('State')
ax.set_title('States with High MH Services Available in 2010')

autolabel(serv10)
plt.savefig('s2010.jpg')
```
